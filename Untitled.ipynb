```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn import datasets, svm,linear_model
import seaborn as sns
import csv
```

```python
# ignore all future warnings
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
#read training & test dataset
df = pd.read_csv(r"C:\Users\User-5\Documents\Python Scripts\Jupyter Notebook Files\titanic competition\train.csv") 
dftest = pd.read_csv(r"C:\Users\User-5\Documents\Python Scripts\Jupyter Notebook Files\titanic competition\test.csv") 
df
```

```python
#replace null values with column mean
df.Fare.fillna(df.Fare.mean(),inplace=True)
df.Age.fillna(df.Age.median(), inplace=True)

dftest = dftest.fillna(df.mean())

#Create new column familysize
df['FamSize'] = df['SibSp'] + df['Parch']
dftest['FamSize'] = dftest['SibSp'] + dftest['Parch']

#Create new column child or adult
#df['Child'] = np.where(df['Age']<=18, 1, 0)
#df['Adult'] = np.where(df['Age']>18, 1, 0)
#dftest['Child'] = np.where(dftest['Age']<=18, 1, 0)
#dftest['Adult'] = np.where(dftest['Age']>18, 1, 0)
```

```python
#use dummy variables for categorical data - split into one column per value with 1s & 0s
d = {1:'1st',2:'2nd',3:'3rd'} #Creating a dictionary to convert Passenger Class from 1,2,3 to 1st,2nd,3rd.
df['Pclass'] = df['Pclass'].map(d) #Mapping the column based on the dictionary
dftest['Pclass'] = dftest['Pclass'].map(d) #Mapping the column based on the dictionary

df = pd.get_dummies(df, columns=["Pclass","Embarked", "Sex"], prefix=["Pclass","Embarked", "Sex"])
dftest = pd.get_dummies(dftest, columns=["Pclass","Embarked", "Sex"], prefix=["Pclass","Embarked", "Sex"])

x = df.drop(["PassengerId","SibSp","Parch","Survived","Name","Ticket","Cabin"],axis=1)
xtest = dftest.drop(["PassengerId","SibSp","Parch","Name","Ticket","Cabin"],axis=1)
```

```python
#Split data into X and Y and then Train and Test data. Drop irrellevant columns
y = df.Survived

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#Import the Logstic Regression Model from sklearn library
lm = linear_model.LogisticRegression()

#fit the model on the training data
model = lm.fit(x_train,y_train)

#calculate prediction on the test data
predictions = lm.predict(x_test)
#predictions_test = lm.predict(xtest)


from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test,predictions))
```

```python
#Import the Gradient Boosting Classifier Model from sklearn library

from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(learning_rate=0.1,max_depth=3)
model.fit(x_train, y_train)
predictions_grad = model.predict(x_test)

from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, predictions_grad))
print(classification_report(y_test,predictions_grad))
```

```python
#calculate model accuracy
score = lm.score(x_test, y_test)
score
```

```python
#Predict on Test Set using Logistic Regression
predictions_test = lm.predict(xtest)
new_pred = pd.DataFrame(predictions_test)
export = np.column_stack((dftest.PassengerId, new_pred))
```

```python
#Predict on Test Set using  Boosting Classifier Model 
predictions_test = model.predict(xtest)
new_pred = pd.DataFrame(predictions_test)
export = np.column_stack((dftest.PassengerId, new_pred))
export
```
